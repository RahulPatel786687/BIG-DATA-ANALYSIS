```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, avg, count
```

```python
spark = SparkSession.builder.appName("BigDataAnalysis").getOrCreate()
```

## Load Dataset

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
!wget -O /tmp/tips.csv $url
df = spark.read.csv("/tmp/tips.csv", header=True, inferSchema=True)
df.show(5)
```

## Basic Analysis

```python
# Total records
df.count()
```

```python
# Average bill
df.select(avg("total_bill")).show()
```

```python
# Customers per day
df.groupBy("day").count().show()
```

```python
# High value bills
df.filter(col("total_bill") > 20).show()
```

```python
# Gender-wise average spending
df.groupBy("sex").avg("total_bill").show()
```

## Scalability Demonstration

```python
df = df.repartition(4)
```

## Insights

- Most customers visit on weekends.
- Average bill value observed.
- Spending patterns differ across gender.
- High-value transactions are fewer.

PySpark enables distributed data processing, making it scalable for large datasets.
